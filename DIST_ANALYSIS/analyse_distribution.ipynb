```python
import os
import pandas as pd
import numpy as np
import altair as alt
root_path=os.getcwd()
case='isic2018_efficientb3_CE_W_None_F_32'
filenames_out='def_pred_distribution_'+case+'.csv'

bars=alt.Chart(filenames_out,
          title='Probabilities distribution per epoch of Dermatofibroma with CE loss'
).mark_circle(
    opacity=0.8,
    stroke='black',
    strokeWidth=1
).encode(
    alt.X('Epoch:Q', axis=alt.Axis(labelAngle=0),sort='ascending'),
    alt.Y('Probabilities:N'),
    alt.Size('Number_of_samples:Q',
        scale=alt.Scale(range=[5, 600]),
        legend=alt.Legend(title='Size')
    ),
    alt.Color('Number_of_samples:Q', scale=alt.Scale(scheme="reds"))
).properties(
    width=2000,
    height=600
)
bars.configure_title(fontSize=34).configure_axis(titleFontSize=24,labelFontSize=24).configure_legend(titleColor='black', titleFontSize=24)
```

```python
filenames_out='nev_pred_distribution_'+case+'.csv'

bars=alt.Chart(filenames_out,
          title='Probabilities distribution per epoch of Nevi with CE loss'
).mark_circle(
    opacity=0.8,
    stroke='black',
    strokeWidth=1
).encode(
    alt.X('Epoch:Q', axis=alt.Axis(labelAngle=0),sort='ascending'),
    alt.Y('Probabilities:N'),
    alt.Size('Number_of_samples:Q',
        scale=alt.Scale(range=[5, 600]),
        legend=alt.Legend(title='Size')
    ),
    alt.Color('Number_of_samples:Q', scale=alt.Scale(scheme="blues"))
).properties(
    width=2000,
    height=600
)
bars.configure_title(fontSize=34).configure_axis(titleFontSize=24,labelFontSize=24).configure_legend(titleColor='black', titleFontSize=24)
```
